```python
import pandas as pd
df=pd.read_csv('shot_logs.csv')
#boş olan satır var mı varsa nerede bulunuyor ve kaç satır boş buna bakyoruz.
df.isnull().sum()
```

```python
#Boş olan satırların olduğu yerleri siliyoruz ve kullanmayacağımız featureları siliyoruz.
df=df.dropna()
df = df.drop('MATCHUP',axis=1)
df = df.drop('CLOSEST_DEFENDER',axis=1)
df = df.drop('player_name',axis=1)
df.dtypes
```

```python
#int ya da float olmayan değerlerimizi int'e çeviriyoruz
df['W'].replace("W",1, inplace=True)
df['W'].replace("L",0, inplace=True)
df['SHOT_RESULT'].replace("made",1, inplace=True)
df['SHOT_RESULT'].replace("missed",0, inplace=True)
df.dtypes
```

```python
df.head()
```

```python
df1 = df['GAME_CLOCK'].str.split(':', expand=True).astype(int)
df['time'] = df1[0] * 60 + df1[1] 
df = df.drop('GAME_CLOCK',axis=1)
df['LOCATION'].replace("A",1, inplace=True)
df['LOCATION'].replace("H",0, inplace=True)
df.head()
```

```python
df.dtypes
```

```python
#dokunma süresinde negatif süreler var mı diye baktık
(df["TOUCH_TIME"]<0).sum()
```

```python
#bulunan negatif değerlerin bulunduğu satırları sildik
df=df[df["TOUCH_TIME"] >=0]
(df["TOUCH_TIME"]<0).sum()
```

```python
X = df.drop('SHOT_RESULT',axis=1)
Y = df['SHOT_RESULT']
from sklearn.model_selection import StratifiedKFold
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
skf = StratifiedKFold(n_splits=10)
skf.get_n_splits(X, Y)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import matthews_corrcoef
from sklearn.metrics import f1_score
```

```python
acclist=[]
precisionscorelist=[]
recallscorelist=[]
mcclist=[]
f1scorelist=[]
i=0
def Average(lst):
    return sum(lst) / len(lst)
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_roc_curve(fpr, tpr, auc_score):
    plt.plot(fpr, tpr, color='orange', label='ROC')
    plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
    plt.xlabel('Yanlış Pozitif Oran')
    plt.ylabel('Doğru Pozitif Oran')
    plt.title('ROC EĞRİSİ'.format(accort))
    plt.legend()
    plt.show()
```

```python
def plot_confusion_matrix(cm, title='Confusion matrix', cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(2)
    plt.xticks(tick_marks, [0,1])
    plt.yticks(tick_marks, [0,1])
    plt.tight_layout()
    plt.ylabel('Beklenen Değer')
    plt.xlabel('Tahmin Edilen Değer')
```

```python
print('Naive Bayes')
for train_index, test_index in skf.split(X, Y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    Y_train, Y_test = Y.iloc[train_index], Y.iloc[test_index]
    model = GaussianNB ()
    model.fit(X_train,Y_train)
    model.score(X_test, Y_test)
    Y_test[:10]
    Y_pred=model.predict(X_test)
    acclist.insert(i,accuracy_score(Y_test, Y_pred))
    precisionscorelist.insert(i,precision_score(Y_test, Y_pred))
    recallscorelist.insert(i,recall_score(Y_test, Y_pred))
    mcclist.insert(i,matthews_corrcoef(Y_test, Y_pred))
    f1scorelist.insert(i,f1_score(Y_test, Y_pred))
accort=Average(acclist)
precisionort=Average(precisionscorelist)
recallort=Average(recallscorelist)
mccort=Average(mcclist)
f1scoreort=Average(f1scorelist)
print("Accuracy",accort)
print("Precisions",precisionort)
print("Recall",recallort)
print("MCC",mccort)
print("F1core",f1scoreort)
```

```python
from sklearn.neighbors import KNeighborsClassifier
print('kNN')
for train_index, test_index in skf.split(X, Y):
    i=0
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    Y_train, Y_test = Y.iloc[train_index], Y.iloc[test_index]
    knn=KNeighborsClassifier(n_neighbors=5,metric='minkowski')
    knn.fit(X_train,Y_train)
    Y_pred=knn.predict(X_test)
    print(confusion_matrix(Y_test,Y_pred))
    acclist[i]=(accuracy_score(Y_test, Y_pred))
    precisionscorelist[i]=(precision_score(Y_test, Y_pred))
    recallscorelist[i]=(recall_score(Y_test, Y_pred))
    mcclist[i]=(matthews_corrcoef(Y_test, Y_pred))
    f1scorelist[i]=(f1_score(Y_test, Y_pred))
    ++i
accort=Average(acclist)
precisionort=Average(precisionscorelist)
recallort=Average(recallscorelist)
mccort=Average(mcclist)
f1scoreort=Average(f1scorelist)
print("Accuracy",accort)
print("Precisions",precisionort)
print("Recall",recallort)
print("MCC",mccort)
print("F1core",f1scoreort)
```
