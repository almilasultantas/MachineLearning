```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("shot_logs.csv")
df.head()
```

```python
#df = df.dropna()
df = df.drop("MATCHUP", axis = 1)
df = df.drop('CLOSEST_DEFENDER',axis=1)
df = df.drop('player_name',axis=1)
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = "mean")
S_C = df.iloc[:, 7:8].values
print(S_C)
imputer= imputer.fit(S_C[:])
S_C[:] = imputer.transform(S_C[:])
print(S_C)
df.head()
```

```python
len(df)
```

```python

sc = pd.DataFrame(data = S_C, index = range(128069), columns = ["shot_clock"])
df = pd.concat([df, sc], axis = 1)
df.head()
```

```python
df = df.drop("SHOT_CLOCK", axis = 1)
```

```python
df["W"].replace("W", 1, inplace = True)
df["W"].replace("L", 0, inplace = True)
df["SHOT_RESULT"]. replace("made", 1, inplace = True)
df["SHOT_RESULT"].replace("missed", 0, inplace = True)
df.dtypes
```

```python
df1 = df["GAME_CLOCK"].str.split(":", expand = True).astype(int)
df["time"] = df1[0] * 60 + df1[1]
df = df.drop("GAME_CLOCK", axis = 1)
df["LOCATION"].replace("A", 1, inplace = True)
df["LOCATION"].replace("H", 0, inplace = True)
df.head()
```

```python
(df["TOUCH_TIME"] < 0).sum()
```

```python
df = df[df["TOUCH_TIME"] >= 0]
(df["TOUCH_TIME"] < 0).sum()
```

```python
X = df.drop("SHOT_RESULT", axis = 1)
Y = df["SHOT_RESULT"]
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
skf = StratifiedKFold(n_splits=10)
skf.get_n_splits(X, Y)
```
